# 1. Convert the app sizes to a number

```python
# Funzione che trasforma una stringa in un numero
def string_to_numeric(string):
    try:
        return int(string)
    except:
        return float(string)

# Funzione che converte la variabile 'Size' in un numero
def convert_size(size):
    if re.search("M", size):
        size = re.sub("M","",size)
        size = int(string_to_numeric(size) * 1000000)
    elif re.search("k", size):
        size = re.sub("k","",size)
        size = int(string_to_numeric(size) * 1000)
    return size

df["Size"] = df["Size"].apply(convert_size)
df.head()
```

#  2. Convert the number of installs to a number

```python
# Funzione che converte la variabile 'Installs' da stringa a numero
def convert_installs(installs):
    installs = re.sub(",","",installs)
    installs = re.sub("\+","",installs)
    installs = int(installs)
    return installs

df["Installs"] = df["Installs"].apply(convert_installs)
df.head()
```

# 3. Transform “Varies with device” into a missing value

```python
# Funzione che trasforma i valori 'Varies with device' nelle diverse variabili in missing value
def transfor_in_na(obj):
    if obj == "Varies with device":
        return None
    else:
        return obj
    
for i in df.columns:
    df[i] = df[i].apply(transfor_in_na)
df.head()
```

# 4. Convert Current Ver and Android Ver into a dotted number (e.g. 4.0.3 or 4.2)

```python
# funzione che converte le variabili che rappresentano le versioni dell'app e del sistema operativo Android
def convert_version(version):
    try:
        convert = re.compile("^(?P<version>(\d+\.)+\d+)")
        version = convert.search(version).group()
        return version
    except: # per non considerare i missing value
        pass

df["Current Ver"] = df["Current Ver"].apply(convert_version)
df["Android Ver"] = df["Android Ver"].apply(convert_version)
df.head()
```

# 8. Create two dataframes: one for the genres and one bridging apps and genres. So that, for instance, the app Pixel Draw - Number Art Coloring Book appears twice in the bridging table, once for Art & Design, once for Creativity

Creazione del dataframe dei generi

```python
genres = []
for genre in df["Genres"]:
    try:
        tmp = genre.split(";")
        for i in tmp:
            if i not in genres:
                genres.append(i)
    except:
        pass
    
df_genres = pd.DataFrame(genres, columns=["Genres"])
df_genres
```

Creazione dataframe con app e genere

```python
df_app_genres = pd.DataFrame(columns = ["App","Genres"])
for i in range(0,len(df)):
    row = df[["App","Genres"]].iloc[i]
    genre_list = row["Genres"].split(";")
    if len(genre_list) > 1:
        for genre in genre_list:
            row["Genres"] = genre
            df_app_genres = df_app_genres.append(row)
    else:
        df_app_genres = df_app_genres.append(row)

index_df_app_genres = pd.Series(list(range(len(df_app_genres)))) 
df_app_genres = df_app_genres.set_index(index_df_app_genres)
df_app_genres
```

# 11. For each app, compute the approximate income, obtain as a product of number of installs and price.

```python
# Funzione che trasforma la variabile 'Price' da stringa a numerica
def price_to_numeric(price):
    try:
        p = re.compile("(?P<price>(\d+\.\d+))")
        price = p.search(price).group()
        price = float(price)
        return price
    except:
        price = float(price)
        return price
    
df["Price"] = df["Price"].apply(price_to_numeric)
df["Income"] = df["Installs"] * df["Price"]
df[["App","Income"]].sort_values(by = "Income", ascending = False)
```
